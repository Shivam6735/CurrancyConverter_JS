```python
attendance_median = df["Attendance"].median()

high_attendance = df[df["Attendance"] > attendance_median]
low_attendance = df[df["Attendance"] <= attendance_median]

print("High Attendance Avg Score:", high_attendance["Exam_Score"].mean())
print("Low Attendance Avg Score:", low_attendance["Exam_Score"].mean())
```

```python
hours_median = df["Hours_Studied"].median()

high_both = df[
    (df["Attendance"] > attendance_median) & 
    (df["Hours_Studied"] > hours_median)
    ]

low_both = df[
    (df["Attendance"] <= attendance_median) & 
    (df["Hours_Studied"] <= hours_median)
]

print("High Attendance + Hight Study Avg Score : ", high_both["Exam_Score"].mean())
print("Low Attendance + Low Study Avg Score : ", low_both["Exam_Score"].mean())
```

```python
# Above two qualities Study_Hour and Attendance matter more for an student performance
#Students who both attend regularly and study above average score nearly 7 marks higher than students who do neither. 
#Consistency in both effort and presence significantly boosts performance.

# Intercation Effect --> Two moderate drivers combining into a stronger outcome
```
